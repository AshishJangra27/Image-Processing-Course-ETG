```python
import cv2 as cv
import numpy as np

import matplotlib.pyplot as plt
```

```python
img = cv.imread('rgb.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
r = img[:,:,2: ]
g = img[:,:,1:2]
b = img[:,:,:1 ]
```

```python
r = np.reshape(r, (300,400))
plt.imshow(r)
```

```python
g = np.reshape(g, (300,400))
plt.imshow(g)
```

```python
b = np.reshape(b, (300,400))
plt.imshow(b)
```

```python
cv.imwrite("R.png",r)
cv.imwrite("G.png",g)
cv.imwrite("B.png",b)
cv.imwrite("RGB.png",img)
```
